# Data Visualization Diagnosis of Diabetes

## Matplotlib

[Matplotlib](http://matplotlib.org/) is the most common charting package, see its [documentation](http://matplotlib.org/api/pyplot_api.html) and [examples](http://matplotlib.org/gallery.html#statistics) for inspiration.

### Bar Plots

```
import matplotlib.pyplot as plt
diagnosis_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DPF', 'Age']
values = [3,78,50,32,88,31,0.248,26]


# Colors: https://matplotlib.org/api/colors_api.html

plt.bar(diagnosis_names, values, label="Diagnosis", color='b')
plt.plot()

plt.xlabel("Diagnosis of Diabetes")
plt.ylabel("Unit")
plt.title("Diabetes Chart Example")
plt.legend()
plt.show()
```

###BMI vs Glucose Bar Plots

```
import matplotlib.pyplot as plt
BMI = [33.6, 23.3, 43.1, 31, 30.5, 38, 30.1]
Glucose = [148, 183, 137, 78, 197, 125, 168]


# Colors: https://matplotlib.org/api/colors_api.html

plt.bar(BMI, Glucose, label='Glucose', color='g')
plt.plot()

plt.xlabel("Glucose")
plt.ylabel("BMI")
plt.title("BMI vs Glucose Bar Plots")
plt.legend()
plt.show()
```

###*Age* vs (Isurin and Glucose and BMI) Bar Plots

```
import matplotlib.pyplot as plt

Age =     [ 10,  20,  30,  40,  50,  60,  70,  80,  90]
Isurin =  [94, 168, 88, 543, 846, 175, 230, 83, 96]
Glucose = [148, 85, 183, 89, 137, 116, 78, 115, 197]
BMI =     [33.6, 22.6, 23.3, 28.1, 43.1, 25.6, 31, 35.3, 37.6]

# Adding legend for stack plots is tricky.
plt.plot([], [], color='r', label = 'Isurin')
plt.plot([], [], color='g', label = 'Glucose')
plt.plot([], [], color='b', label = 'BMI')

plt.stackplot(Age, Isurin, Glucose, BMI, colors= ['r', 'g', 'b'])
plt.title('Age vs (Isurin and Glucose and BMI) ')
plt.xlabel('Age')
plt.legend()
plt.show()
```

### Line Plots

```
import matplotlib.pyplot as plt

x = [40, 50, 60, 70, 80, 90, 100, 110,120]
Glucose = [115, 113, 117, 125, 103, 105, 95, 104, 108]
BloodPressure = [100,106, 104, 86, 80, 78, 90, 86, 90]
plt.plot(x, Glucose, label="Glucose")
plt.plot(x, BloodPressure, label="Blood Pressure")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Glucose vs Blood Pressure Example")
plt.legend()
plt.show()
```
